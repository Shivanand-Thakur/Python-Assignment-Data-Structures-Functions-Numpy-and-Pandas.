```python
# Section A: Data Structures & Control Structures
# Q1 Solution
nums = [10, 20, 30, 40, 50]

# append
nums.append(60)
print("After append:", nums)

# extend
nums.extend([70, 80])
print("After extend:", nums)

# insert
nums.insert(2, 25)
print("After insert:", nums)

# remove
nums.remove(40)   # removes first occurrence
print("After remove:", nums)

# pop
nums.pop()
print("After pop:", nums)

# index
print("Index of 30:", nums.index(30))

# count
print("Count of 20:", nums.count(20))

# sort
nums.sort()
print("After sort:", nums)

# reverse
nums.reverse()
print("After reverse:", nums)

# clear
nums.clear()
print("After clear:", nums)
```

```python
## Q2. Create a tuple that stores 3 student names. Try changing the second name in the tuple. What happens? Explain why.
# Q2 Solution
students = ("Alice", "Bob", "Charlie")
print("Original tuple:", students)

# Attempt to change (will cause error)
try:
    students[1] = "David"
except TypeError as e:
    print("Error:", e)

# Explanation
print("Tuples are immutable, meaning their elements cannot be changed after creation.")
```

```python
## Q6. Given 'marks', print the grade.
# Q6 Solution
marks = 82  # Change value to test

if marks >= 90:
    grade = "A"
elif marks >= 75:
    grade = "B"
elif marks >= 50:
    grade = "C"
else:
    grade = "Fail"

print("Marks:", marks, "Grade:", grade)
```

```python
## Q7. Given a number, check if it's positive and even. If not positive, print if it's zero or negative.
# Q7 Solution
num = -5  # Change value to test

if num > 0:
    if num % 2 == 0:
        print(num, "is positive and even.")
    else:
        print(num, "is positive but odd.")
elif num == 0:
    print("The number is zero.")
else:
    print(num, "is negative.")
```

```python
# Section C: Exploring Pandas
## Q11. Create a small DataFrame with 10 rows and columns: 'Name', 'Age', 'City', 'Salary'.
## Then perform the following:
##- .info(), .describe()
##- Select 'Name' and 'City' columns
##- Drop 'City' column
##- Fill missing values in 'Salary' with mean
##- Remove duplicate rows
# Q11 Solution

data = {
    "Name": ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"],
    "Age": [22, 25, 23, 28, 30, 26, 22, 25, 29, 24],
    "City": ["Delhi", "Mumbai", "Chennai", "Delhi", "Kolkata", "Mumbai", "Delhi", "Bangalore", "Chennai", "Delhi"],
    "Salary": [50000, 60000, None, 55000, 58000, None, 60000, 65000, 70000, 60000]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Info and Describe
print("\nInfo:")
print(df.info())
print("\nDescribe:")
print(df.describe())

# Select columns
print("\nSelected columns (Name, City):\n", df[['Name', 'City']])

# Drop City
df_no_city = df.drop(columns=['City'])
print("\nAfter dropping City:\n", df_no_city)

# Fill missing salary with mean
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print("\nAfter filling missing Salary:\n", df)

# Remove duplicates
df_no_duplicates = df.drop_duplicates()
print("\nAfter removing duplicates:\n", df_no_duplicates)
```
